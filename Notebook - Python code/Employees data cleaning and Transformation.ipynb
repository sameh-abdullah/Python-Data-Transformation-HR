## Employee Data Cleaning & Transformation ETL Pipeline using python

### Overview

This project demonstrates an ETL (Extract, Transform, Load) pipeline implemented in Python for cleaning, standardizing, and enriching employee data from the csv source.


##### Importing the main package for data transformation
- Panda
- Numpy

```python
import pandas as pd
import numpy as np
import datetime as dt
```

 Extracting data from csv source, full load method

```python
data_url=r'.\Dataset\Messy_Employee_dataset.csv'
employees_df = pd.read_csv(data_url)
```

Display all the main information about the data for starting data cleaning

```python
employees_df.info()
```

Check the data frame columns and clean it (Strip the spaces and change them to lower cases)


```python
cols = employees_df.columns.str.strip().str.lower()

employees_df.columns=cols
employees_df.head(2)
```

Check the uniquness of the employee_id column and make it upper

```python
employees_df.employee_id.is_unique
employees_df.employee_id = employees_df['employee_id'].str.upper()
employees_df.head(2)
```

Remove leading/trailing spaces from names and capitalizing them

```python
name=['first_name','last_name']
for col in name:
    employees_df[col] = employees_df[col].str.strip().str.capitalize()

employees_df.head(3)
```

Split Department_Region into:
                - Department
                - Region

            Example:
                IT-NorthAmerica
                -> IT
                -> NorthAmerica

```python
employees_df['department']=(employees_df['department_region'].astype(str).str.split('-').str[0]
            .str.capitalize())
employees_df['region']=(employees_df['department_region'].astype(str).str.split('-').str[1]
            .str.capitalize())

employees_df.sample(5)
```

Convert employment status into binary flag

            1 = Active
            0 = Inactive / Pending

```python
try:
    employees_df['status'] = employees_df['status'].apply(lambda x: 1 if str.lower(x) in ['active','pending'] else 0)
except Exception as e:
    print(e)
```

Check Join_date column is as date or not, if not we will change it type to date

```python
employees_df.info()
```

```python
employees_df['join_date'].isna().sum()
```

```python
employees_df['join_date'] = pd.to_datetime(employees_df['join_date'])
employees_df.sample(5)
```

Checking emails uniqueness and at the same time has the same domain

```python
employees_df['email'] = employees_df['email'].str.strip()
employees_df[employees_df['email'].duplicated()==True]['email'].count()
# the test data quality shows that there are 956 emial is duplicate
# as action we will flag the as a true or false in new column for check it with the stakeholder
```

```python
employees_df['email'].apply(lambda x : True if '@example.com' in x else False).sum()
# all the email has the same domain
```

```python
employees_df['duplicate_email_flag'] = employees_df['email'].duplicated()
employees_df.sample(5)
```

```python
employees_df.info()
```

Phone numbers are identifiers, not numeric values.
Casting to decimal and remove minuse only.

```python
employees_df['phone'] = pd.to_numeric(employees_df['phone'].astype(str).str.replace('-',''))
employees_df['phone'].sample(5)
```

```python
employees_df['performance_score']=employees_df['performance_score'].str.capitalize()
# in employees_df['performance_score'], every thing is code
# we can standerize the performance_score to 1,2,3,4 value for optimazing the speed of reading
```

Casting the remore_work column to int for data profermance

```python
employees_df['remote_work']=employees_df['remote_work'].astype(int)
# column already boolean type so we decide to convert it to int for data profermance
```

```python
employees_df.info()
```

Handling the missing values in the age and salary columns using the KNN algorithm using the sklearn package
in this algorithm will suppose the neighbor is 5 and weigh is distance to find the nearst k value and fill the
the missing value

```python
from sklearn.impute import KNNImputer
# pip install sklearn
```

```python
numeric_columns=['age','salary'] # split the numeric columns
categories_columns=['employee_id', 'first_name', 'last_name', 'department_region',
       'status', 'join_date', 'email', 'phone', 'performance_score',
       'remote_work', 'department', 'region', 'duplicate_email_flag'] # split the categories columns

numeric_df=employees_df[numeric_columns]
categories_df=employees_df[categories_columns]
```

```python
imputer = KNNImputer(n_neighbors=5, weights='distance') # initialize the knnimputer
```

```python
df_imputed = imputer.fit_transform(numeric_df) # learn the pattern and fill the missing values
```

```python
df_imputed=pd.DataFrame(df_imputed,columns=numeric_columns) # convert the imputed data to dataframe with the
# numeric columns
```

```python
employees_df=pd.concat([categories_df,df_imputed], axis=1) 
# combine the new imputed dataframe with the categorical dataframe to get the finall dataframe with filled
# missing values
```

Reorder the employees_df to the desire order and then remove the unwanted columns

```python
desire_columns_order=['employee_id', 'first_name', 'last_name', 'age', 'region','department',
       'join_date', 'salary', 'email', 'phone','status','performance_score',
       'remote_work', 'duplicate_email_flag']
employees_df = employees_df.reindex(columns=desire_columns_order)
```

```python
employees_df['age']=employees_df['age'].astype(int)
```

Re-write the cleaned and transformed data to the csv file and i add another method to re-write the 
data to sql server as another way and show case

```python
employees_df.to_csv(r'.\Dataset\Employees_cleaned_dataset.csv', index=False)
```
